## Model Building(Simple Linear Regression)

y(Sales) = c + (TV) * x

```python
X = advertising['TV']
y = advertising['Sales']
```

### Train/Test Split
We will split the data into 70%(training) and 30%(testing)

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, test_size=0.3, random_state=100)
```

```python
X_train
```

```python
y_train
```

## Building a linear model

```python
import statsmodels.api as sm
```

```python
# Add a constant to get an intercept
X_train_sm = sm.add_constant(X_train)

# Fit the regression line using OLS
lr = sm.OLS(y_train, X_train_sm).fit()
```

```python
lr.params
```

Sales = 6.94 + 0.054546x

```python
lr.summary()
```

The main values of concern are:

1. Coefficients and significance(p-values):
The coefficients are statistically significant. The association is not by random chance
2. R-Squared
81.6% of the variance in Sales is explained by TV

3. F-Statistic and its significance
The p-value is low(1.52e-52), the model is statistically significant

The equation becomes:
Sales = 6.9487 + 0.0545(TV)

```python
plt.scatter(X_train, y_train)
# Plot the regression line
plt.plot(X_train, 6.9487 + 0.0545*X_train, 'red')
plt.show()
```

## Model Evaluation
### Distribution of the residuals(error terms)

```python
y_train_pred = lr.predict(X_train_sm)
residuals = (y_train - y_train_pred)
```

```python
fig = plt.figure()
sns.displot(residuals, bins=15, kde=True)
fig.suptitle('Error Terms', fontsize=15) # plot heading
plt.xlabel('y_train - y_train_pred', fontsize=15) # X-label
plt.show()
```

Residuals are normally distributed with a mean of 0

### Look for patterns in the residuals

```python
plt.scatter(X_train, residuals)
plt.show()
```

We are confident that the model fit is not by chance. The normality of the residual terms allows some inference on the coefficients

Although, the variance of residuals increasing with X indicates that there is significant variation that this model is unable to explain

## Predictions on the Test set

```python
X_test_sm = sm.add_constant(X_test)
y_pred = lr.predict(X_test_sm)
```

```python
y_pred.head()
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
# Returns the mean squared error: we'll take a square root
np.sqrt(mean_squared_error(y_test, y_pred))
# RMSE: the standard deviation of the resisuals, how far the residuals are concentrated around the
# line of best fit
```

```python
### R-Squared - test set
r_squared = r2_score(y_test, y_pred)
r_squared
```

```python
# Visualising the fit on the test set
plt.scatter(X_test, y_test)
plt.plot(X_test, 6.948+0.054*X_test, 'red')
plt.show()
```
